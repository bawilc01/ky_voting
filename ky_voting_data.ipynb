```python
# Kentucky Registered Voters Data
# Stats: https://elect.ky.gov/Resources/Pages/Registration-Statistics
# Data from November 2020
# xls data transformed imported as tab delimited csv file; removed county code that is appearing as index now


import pandas as pd
import configparser

# import tab delimited file
doc = 'kentucky_vote_info_by_county.csv'
df1 = pd.read_csv(doc, delimiter='\t')


# converting all caps county to first letter cap, rest letter lower case
df1['County']= df1['County'].str.title()


# display data in table; used existing headers from source
from IPython.display import display
display(df1)
```

```python
import json

# set json file location to data var
data = 'ky_total_pop_by_county.json'

df2 = pd.read_json(data)

from IPython.display import display
display(df2)
```

```python
# Merging two dataframes together on county name
# Voting data and population data in same dataframe

df_merge = pd.merge(df1, df2, left_on='County', right_on='County', how='left')

from IPython.display import display
display(df_merge)
```
